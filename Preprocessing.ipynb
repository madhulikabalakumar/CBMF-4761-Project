```python
import anndata as ad
from scipy.sparse import csr_matrix
adata = ad.AnnData(
    X=csr_matrix(matrix),  # Assuming `matrix` is a sparse matrix
    obs=pd.DataFrame(index=rows[0]),  # Observation metadata (e.g., cell information)
    var=pd.DataFrame(index=cols[0])  # Variable metadata (e.g., gene information)
)
```

```python
print(adata.obs)
print(adata.var)
print(adata.X)
```

```python
adata.obs.index = adata.obs.index.astype(str)
adata.var.index = adata.var.index.astype(str)

# Optionally, add a check to confirm the indices are correctly set
print("Initial obs indices:", adata.obs.index[:5])
print("Initial var indices:", adata.var.index[:5])
```

```python
print("Current obs index name:", adata.obs.index.name)
print("Current var index name:", adata.var.index.name)

# Set index names to None if they are not meaningful
adata.obs.index.name = None
adata.var.index.name = None

# Save the AnnData object
file_path = '/content/drive/MyDrive/computational genomic/Genomics Project Data/Aggregated_filtered_normalised_count.h5ad'
adata.write(file_path)

# Load the AnnData object from the saved file
adata_loaded = ad.read_h5ad(file_path)

# Check indices after loading to confirm they are preserved correctly
print("Loaded obs indices:", adata_loaded.obs.index[:5])
print("Loaded var indices:", adata_loaded.var.index[:5])
```

```python
print("Data matrix shape:", adata_loaded.X.shape)

# If there is additional metadata, check its preservation as well
if hasattr(adata_loaded, 'uns'):
    print("Metadata in 'uns':", adata_loaded.uns.keys())

# Ensure 'obsm' and 'varm' are also intact if used
if hasattr(adata_loaded, 'obsm'):
    print("Keys in 'obsm':", adata_loaded.obsm.keys())
if hasattr(adata_loaded, 'varm'):
    print("Keys in 'varm':", adata_loaded.varm.keys())
```
